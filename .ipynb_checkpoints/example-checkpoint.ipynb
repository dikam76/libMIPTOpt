```python
import numpy as np
import grad as g
```

```python
n= 10
A=np.eye(n)*np.arange(1,11)
b=np.ones(n)
x0=np.zeros(n)
```

```python
def func(A,b,x):
    return 0.5*x.T @ A @ x - b.T @ x
def grad(A,b,x):
    return A @ x - b
```

```python
k=100
xk=x0

for i in range(k):
    xk=g.gradient_step(xk, grad(A, b, xk), A)
    print(xk)
    print(func(A,b,xk))
```
