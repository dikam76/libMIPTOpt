```python
import numpy as np
import gd as g
import matplotlib.pyplot as plt
```

```python
n = 10
A = np.eye(n)*np.arange(1,11)
b = np.ones(n)
x0 = np.zeros(n)
```

```python
func = lambda x: 0.5 * (A @ x - b).T @ (A @ x - b)
grad = lambda x: A.T @ ( A @ x - b)
```

```python
def Heavy_ball(x0, func, grad, t):
    xk, xk_prev = x0, x0
    res = [func(xk)]
    for i in range(t):
        xk_t = g.fast_grad_des_v1(xk, xk_prev, grad, 0.01)
        xk_prev = xk
        xk = xk_t
        res.append(func(xk))
    return xk, res
```

```python
xk, res = Heavy_ball(x0, func, grad, 2000)
```

```python
plt.plot(np.log(res))
plt.show()
```
