```
import matplotlib.pyplot as plt
import numpy as np
```

```
np.random.seed(1)
#print(os.getcwd())
u_orig = plt.imread("роза")
rows, cols, colors = u_orig.shape
print (u_orig[65][78])
print (cols)
known = np.zeros((rows, cols, colors))
for i in range(rows):
    for j in range(cols):
        if np.random.random() > 0.000001:
            for k in range(colors):
                known[i, j, k] = 10
u_corr = known * u_orig
#print(u_corr)
```

```
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(u_orig, cmap='gray');
ax[0].set_title("Original Image")
ax[0].axis('off')
ax[1].imshow(u_corr);
ax[1].set_title("Corrupted Image")
ax[1].axis('off');
```

```
u_rec = u_corr.copy()
u_grad = np.zeros((rows, cols, 4))
```

```
#norm2
h = 0.1
e = 1e-9
for k in range(100):
    for i in range(1, rows - 1):
        for j in range(1, cols - 1):
            grad1 = 2 * u_corr[i][j] - u_corr[i][j + 1] - u_corr[i + 1][j]
            grad2 = (u_corr[i][j] - u_corr[i][j + 1])**2 + (u_corr[i][j] - u_corr[i + 1][j])**2 + e
            grad3 = 2 * (u_rec[i][j] - u_corr[i][j])
            grad = grad1 / grad2**0.5 + grad3 * h / rows / cols

            #grad1 = u_corr[i][j] / ((u_corr[i][j]**2 + u_corr[i][j + 1]**2 + e)**0.5) 
            #grad2 = u_corr[i][j] / ((u_corr[i][j]**2 + u_corr[i + 1][jrec]**2 + e)**0.5)
            #grad3 = u_corr[i][j] / ((u_corr[i - 1][j]**2 + u_corr[i][j]**2 + e)**0.5)
            #grad4 = u_corr[i][j] / ((u_corr[i][j - 1]**2 + u_corr[i][j]**2 + e)**0.5)
            #grad5 = 2 / (rows * cols * colors) * (u_corr1[i][j] - u_corr[i][j])
            u_rec[i][j] = u_rec[i][j] - h * grad
            #rint('DBG')
```

```
#norm1
h = 0.01
for i in range(1, rows - 1):
    for j in range(1, cols - 1):
        tmp = np.array([1 if u_corr[i][j][k] >= u_corr[i][j + 1][k] else -1 for k in range(colors)]) 
        tmp += np.array([1 if u_corr[i][j][k] >= u_corr[i + 1][j][k] else -1 for k in range(colors)])
        u_rec[i][j] += tmp * h
            
```

```
#norm2_2
h = 0.001
e = 1e-9
for k in range(1):
    for i in range(1, rows - 1):
       for j in range(1, cols - 1):
          for k in range(colors):
              grad1 = (2 * u_rec[i][j][k] - u_rec[i + 1][j][k] - u_rec[i][j + 1][k])
              grad2 = ((u_rec[i][j][k] - u_rec[i][j+1][k]))**2 
              grad0 = (u_rec[i][j][k] - u_rec[i + 1][j][k])**2 + e
              grad = (u_corr[i][j][k] - u_rec[i][j][k]) / rows / cols + grad1 / (grad2 + grad0)**0.5
              #print(grad)
              grad3 = (u_rec[i][j][k] - u_rec[i][j + 1][k])**2 + e + (u_rec[i][j][k] - u_rec[i + 1][j][k])**2
              u_grad[i][j][k] = grad 
              u_grad[i + 1][j][k] = (u_rec[i][j][k] - u_rec[i + 1][j][k]) / grad3**0.5
              u_grad[i][j + 1][k] = (u_rec[i][j][k] - u_rec[i][j + 1][k]) / grad3**0.5
```

```
for k in range(1000):
    for i in range(1, rows - 1):
        for j in range(1, cols - 1):
          u_rec[i][j] = u_rec[i][j] - h * u_grad[i][j]
```

```
fig, ax = plt.subplots(1, 2, figsize = (10, 5))
ax[0].imshow(u_rec)
ax[0].set_title("In-Painted Image")
ax[0].axis('off')
ax[1].imshow(u_corr);
ax[1].set_title("Corrupted Image")
ax[1].axis('off');
```
